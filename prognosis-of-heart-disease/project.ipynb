Согласно медицинской статье, восходящая депрессия сегмента ST обычно считается нормой, в то время как пологая или нисходящая депрессия сегмента ST может указывать на наличие острой ишемии, которая является типом сердечного заболевания.

- У значительного числа пациентов с плоской и наклонной депрессией сегмента ST имеется порок сердца.

## One-hot encoding

One-hot encoding - процесс, с помощью которого категориальные переменные преобразуются в подходящую ​​алгоритмам Машинного обучения форму. Входные данные для этого преобразователя должны представлять собой массив целых чисел или строк, обозначающий значения, принимаемые категориальными (дискретными) признаками. Функции кодируются с использованием схемы кодирования one-hot (также известной как «one-of-K» или «dummy»). Метод создает n новых Features, где n — количество различных категорий в исходной функции. Новые объекты представлены в виде бинарных векторов, где соответствующему признаку присваивается значение 1, если исходное значение соответствует этой категории, и 0, если нет.

Мы будем использовать One-hot encoding для ChestPainType, RestingECG и ST_Slope, поскольку эти функции имеют более двух различных категорий.

В отличие от One-hot encoding, Label encoding не создает новые столбцы. Вместо этого он присваивает каждой метке уникальное целое число в алфавитном порядке.

##Логистическая регрессия

Логистическая регрессия — это алгоритм классификации в машинном обучении для прогнозирования вероятности категориально зависимой переменной.  Она получила свое название благодаря тому, что использует логистическую функцию для прогнозирования вероятности принадлежности объекта к одному из классов.
Логистическая регрессия использует линейную комбинацию входных признаков и соответствующих весов, которая описывает линейную гиперплоскость в пространстве признаков. Затем этот результат проходит через логистическую функцию, которая переводит линейную комбинацию в вероятность принадлежности объекта к одному из классов.

linear-regression-vs-logistic-regression.png

По своей сути логистическая регрессия просто берет уравнение линейной регрессии и использует его как параметр сигмовидной функции. Математически это выражается следующим образом:



f.png

где:

Y — бинарный выходной результат (0 или 1)

X — вектор признаков, используемый для прогнозирования $Y$

P(Y=1|X) — вероятность того, что $Y$ равно 1 при заданном $X$

beta_0, beta_1, beta_2, …, beta_p — коэффициенты модели, которые нужно определить в ходе обучения, чтобы достичь наилучшего соответствия данных

e — число Эйлера

Логистическая регрессия также может быть использована для многоклассовой классификации, когда необходимо разделить данные на более чем два класса. Для этого обучают K моделей, каждая из которых отличается только целевым классом. По сути задача бинарной классификации решается несколько раз и выдается совокупное решение нескольких моделей.

В целом, логистическая регрессия — это мощный инструмент для решения задач бинарной и многоклассовой классификации в Python. Она проста в использовании и предоставляет множество метрик для оценки качества работы модели.



В данном методе мы использовали следующие гиперпараметры:

- penalty: штраф на логистическую модель за слишком большое количество переменных

- solver: решатель, который пытается найти веса параметров, которые минимизируют функцию стоимости

- С: Обратная сила регуляризации

Для получения наивысшей точности работы модели, необходимо правильно подобрать гиперпараметры. Изменение параметров модели может принципиально повлиять на ее качество. Например, модель может переобучиться. Перебор этих параметров вручную может занять колоссальное количество времени.
Поэтому в своей работе мы использовали инструмент GridSearchCV.
.

> GridSearchCV — это, по сути, метод перекрестной проверки для поиска оптимальных значений гиперпараметров из заданного набора параметров в сетке
«Лучшие» параметры, которые идентифицирует GridSearchCV, являются технически лучшими, которые можно было бы создать, но только по параметрам, которые вы включили в свою сетку параметров. Grid Search CV систематически работает с несколькими комбинациями настроек параметров, проверяет каждую из них и определяет, какая из них дает наилучшие характеристики.

## Дерево решений

Дерево
решений - это алгоритм контролируемого машинного обучения, который
идеально подходит для задач классификации, так как он способен
упорядочивать классы на точном уровне. Он работает как блок-схема,
разделяя точки данных на две похожие категории одновременно от
“ствола дерева” до “ветвей” и “листьев”, где категории становятся более
похожими. Это создает категории внутри категорий, позволяя проводить
органическую классификацию с ограниченным наблюдением со
стороны человека.

Дерево решений представляет собой древовидную структуру, похожую на блок-схему, где каждый внутренний узел обозначает проверку атрибута, каждая ветвь представляет результат проверки, а каждый конечный узел содержит метку класса.В дереве решений есть два узла: узел решения и конечный узел. Узлы решений используются для принятия любого решения и имеют несколько ветвей, тогда как конечные узлы являются результатом этих решений и не содержат никаких дополнительных ветвей.
decision-tree-classification-algorithm.png

В объекте Decision Tree от Scikit-learn есть несколько гиперпараметров.

Гиперпараметры, которые мы будем использовать и исследовать здесь:

- criterion: Функция для измерения качества разделения. Поддерживаемые критерии: «gini», «log_loss» и «entropy» для прироста информации Шеннона.

- min_samples_leaf: Минимальное количество выборок, необходимое для конечного узла. Точка разделения на любой глубине будет рассматриваться только в том случае, если она оставляет по крайней мере min_samples_leaf обучающие выборки в каждой из левой и правой ветвей. Это может иметь эффект сглаживания модели, особенно в регрессии.

- max_depth: максимальная глубина дерева. Выбор более низкого значения max_depth может уменьшить количество расщеплений и уменьшить переобучение.

- max_leaf_nodes: Лучшие узлы определяются как относительное уменьшение примесей. Если нет, то неограниченное количество листовых узлов.

## KNN

Метод k-ближайших соседей (k-nearest neighbors) – это простой алгоритм машинного обучения с учителем, который можно использовать для решения задач классификации и регрессии.

Алгоритм K-ближайших соседей — это непараметрический классификатор обучения с учителем, который классифицирует данные или предсказывает принадлежность их к классам на основе близости, то есть алгоритм находит расстояния между запросом и всеми примерами в данных, выбирая определенное количество примеров (k), наиболее близких к запросу.

k-nearest-neighbor-algorithm-for-machine-learning2.png


В случае использования метода для классификации объект присваивается тому классу, который является наиболее распространённым среди k соседей данного элемента, классы которых уже известны.
В случае использования метода для регрессии, объекту присваивается среднее значение по k ближайшим к нему объектам, значения которых уже известны.

**Алгоритм:**

1. Задать количество соседей k.
2. Измерить евклидово расстояние между объектом, для которого предсказывается класс, и всеми объектами в обучающей выборке.
3. Выбрать k объектов с наименьшими расстояниями.
4. Определить класс предсказываемого образца как класс, который преобладает среди k выбранных объектов.

Используемые гиперпараметры:

- n_neighbors: количество соседей

## GradientBoostingClassifier

GradientBoostingClassifier — это техника машинного обучения для задач классификации и регрессии, которая строит модель предсказания в форме ансамбля слабых предсказывающих моделей, обычно деревьев решений.

В отличие от случайного леса, градиентный бустинг строит последовательность деревьев, в которой каждое дерево пытается исправить ошибки предыдущего. По умолчанию в градиентном бустинге деревьев регрессии отсутствует случайность, вместо этого используется строгая предварительная обрезка. В градиентном бустинге деревьев часто используются деревья небольшой глубины, от одного до пяти уровней, что делает модель меньше с точки зрения памяти и ускоряет вычисление прогнозов.Gradient Boosting обучает многие модели постепенно, аддитивно и последовательно.

Основная идея градиентного бустинга заключается в объединении множества простых моделей, деревьев небольшой глубины. Каждое дерево может дать хорошие прогнозы только для части данных и таким образом для итеративного улучшения качества добавляется все большее количество деревьев.

Flow-chart-of-XGBoost.png

Используемые гиперпараметры:

- learning_rate: контролирует, насколько сильно каждое дерево будет пытаться исправить ошибки предыдущих деревьев

- n_estimators: Количество этапов бустинга(Добавление количества деревьев )

- criterion: Функция измерения качества разделения

- loss: Функция потерь

- max_features: особенности, которые следует учитывать при поиске лучшего сплита

- max_depth: Максимальная глубина отдельных регрессионных оценок. Максимальная глубина ограничивает количество узлов в дереве.

## SVМ

Метод опорных векторов ( support vector machine, SVM) — метод обучения, который применяется для решения задач классификации и регрессии. Основная идея метода заключается в построении гиперплоскости, разделяющей объекты выборки оптимальным способом. SVM это алгоритм, который получает на входе данные, и возвращает такую разделяющую линию.

Алгоритм SVM устроен таким образом, что он ищет точки на графике, которые расположены непосредственно к линии разделения ближе всего. Эти точки называются опорными векторами. Затем, алгоритм вычисляет расстояние между опорными векторами и разделяющей плоскостью. Это расстояние которое называется зазором. Основная цель алгоритма — максимизировать расстояние зазора.Таким образом меньше будет средняя ошибка классификатора. Лучшей гиперплоскостью считается такая гиперплоскость, для которой этот зазор является максимально большим.

- gamma: Коэффициент ядра
